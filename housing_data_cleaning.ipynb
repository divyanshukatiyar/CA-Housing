```python
import pandas as pd
import missingno as msng
```

```python
housing = pd.read_csv('housing.csv')
housing.head()
```

```python
msng.matrix(housing)
```

```python
housing.isnull().sum()
```

```python
# Since it doesn't make sense for the number of total bedrooms to have a float value, we take median instead 
#of mean to avoid float values
housing['total_bedrooms'].fillna((housing['total_bedrooms'].median()), inplace=True)
```

```python
msng.matrix(housing)
```

```python
housing['ocean_proximity'].value_counts()
```

```python
housing.to_csv('housing_cleaned.csv')
```
